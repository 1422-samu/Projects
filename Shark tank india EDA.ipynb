```python
ashneer=list(df.ashneer_present)
anupam=list(df.anupam_present)
aman=list(df.aman_present)
namita=list(df.namita_present)
vineeta=list(df.vineeta_present)
peyush=list(df.peyush_present)
ghazal=list(df.ghazal_present)


x=pd.DataFrame({'Sharks':['ASHNEER','ANUPAM','AMAN','NAMITA','VINEETA','PEYUSH','GHAZAL'],
              'Number_of_appearance':[sum(ashneer),sum(anupam),sum(aman),sum(namita),sum(vineeta),sum(peyush),sum(ghazal)]})
plt.figure(figsize=(7,7))
sns.barplot(x=x.Sharks,y=x.Number_of_appearance,data=x);
```

# Anupam appeared the most while Gazal appeared least

```python
ashneer=list(df.ashneer_deal)
anupam=list(df.anupam_deal)
aman=list(df.aman_deal)
namita=list(df.namita_deal)
vineeta=list(df.vineeta_deal)
peyush=list(df.peyush_deal)
ghazal=list(df.ghazal_deal)
x=pd.DataFrame({'Sharks':['ASHNEER','ANUPAM','AMAN','NAMITA','VINEETA','PEYUSH','GHAZAL'],
              'Number_of_Deals':[sum(ashneer),sum(anupam),sum(aman),sum(namita),sum(vineeta),sum(peyush),sum(ghazal)]})
plt.figure(figsize=(7,7))
sns.barplot(x=x.Sharks,y=x.Number_of_Deals,data=x);
```
